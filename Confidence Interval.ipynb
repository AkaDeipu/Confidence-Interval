# Importing libraries and data

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
n = len(data) # finding the length of the data
n
```

```python
#Calculating the mean and standard deviation of the sample
sample_mean = np.mean(data)
sample_std = np.std(data, ddof = 1)
print(f"Sample Mean = {sample_mean}")
print (f"Sample Standard Deviation = {sample_std}")
```

```python
pd.DataFrame(data).describe() # Analyzing the data
```

```python
#Visualizing the data with box plot
pd.DataFrame(data).plot(kind = 'box', figsize = (7,4), title = "Box Plot of print heads", showmeans = True ,
                        meanline = True, notch = True, patch_artist = True, color = 'blue')
plt.tight_layout()
plt.show()
```

This box plot tells that most the data concentrated between 1.18 and 1.36 suggesting most print heads fail within this range. There's one clear outlier below the lower whisker, around 0.85. This print-head has significantly lower durability compared to others. The mean and median were almost at the same level whic tells that the impact of outliers is not much on the data.

# Performing the T-test

```python
#As the population size is 15 and population standard deviation is not known we perform the t-test
#Calculating the t values with 99% confidence and df as n-1
t_test = stats.t.ppf(0.99 ,df = n-1)
t_test
```

```python
#Calculating the margin of error
margin_error  = t_test * (sample_std / np.sqrt(n))
```

```python
#Calculating the confidence interval range
ci = np.round(sample_mean - margin_error,3) , np.round(sample_mean + margin_error,3)
print (f"Confidence Interval = {ci}")
```

# Performing the Z-test

```python
population_std = 0.2 #Population standard deviation is known so we are performing the z test
```

```python
z_val = stats.norm.ppf(0.99) #Calculating the z value with 99% confidence
z_val
```

```python
margin_error_z = z_val * (population_std / np.sqrt(n)) # Calculating the argin of error
```

```python
#Calculating the confidence interval range
ci_z = np.round(sample_mean - margin_error_z,3) , np.round(sample_mean + margin_error_z,3)
print (f"Confidence Interval for z test = {ci_z}")
```
